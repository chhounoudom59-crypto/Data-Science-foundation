#### 1. Understanding the Dataset

**Objective:** Familiarize yourself with the dataset's structure and the type of information it contains.

    Tasks:
    - Identify the variables and their data types.
    - Determine the number of entries in the dataset.
    - Check for missing values.

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
````

````python
df = pd.read_csv('Salaries.csv')
df
````

````python
#Identify the variables and their data types.
df.info()
````

````python
#Determine the number of entries in the dataset.
print(df.columns)
print(df.shape)
````

````python
#Check for missing values.
df.isnull().sum()
````

#### 2. Data Cleaning
    - Handle missing values, if any.

````python
df.dropna(inplace=True)
df.drop_duplicates(inplace=True)
````

#### 3. Exploratory Data Analysis (EDA)
**Objective:** Explore the data to uncover initial insights and identify patterns.

    Tasks:
    - Analyze the distribution of salaries across different job titles(just three type such as ```['Data Scientist', 'Data Engineer', 'Data Analyst']), experience levels, and company sizes.
    - Explore the relationship between remote work ratio and salary.
    - Investigate trends over the work years provided in the dataset.

````python
dist_df = df[(df['job_title']=='Data Scientist') | (df['job_title']=='Data Engineer') | (df['job_title']=='Data Analyst')]
dist_df
````

````python
for jt in dist_df['job_title'].unique():
    fig, ax = plt.subplots(figsize = (12,6))
    sns.histplot(dist_df[dist_df['job_title']==jt], x='salary_in_usd', hue='experience_level', bins=50)
    ax.set_xlabel("Salary of " + jt)
    ax.set_title("Salary Distribution by Experience Level")
    plt.show()
for jt in dist_df['job_title'].unique():
    fig, ax = plt.subplots(figsize = (12,6))
    sns.histplot(dist_df[dist_df['job_title']==jt], x='salary_in_usd', hue='company_location', bins=50)
    ax.set_xlabel("Salary of " + jt)
    ax.set_title("Salary Distribution by Company Location")
    plt.show()
````

````python
for jt in dist_df['job_title'].unique():
    fig, ax = plt.subplots(figsize = (12,6))
    sns.histplot(dist_df[dist_df['job_title']==jt], x='salary_in_usd', hue='remote_ratio', bins=50)
    ax.set_xlabel("Salary of " + jt)
    ax.set_title("Relationship between Remote Work Ratio and Salary of " + jt)
    plt.show()
````

````python
dist_df['work_year'] = dist_df['work_year'].astype(str)
fig, ax = plt.subplots(figsize = (12,6))
for jt in dist_df['job_title'].unique():
    d_df = dist_df[dist_df['job_title']==jt].groupby('work_year')['job_title'].count().to_frame()
    sns.lineplot(x=d_df.index, y=d_df['job_title'], label=jt)

ax.set_xlabel("Work Year")
ax.set_ylabel("Number of Worker")
ax.set_title("Trends Over the Work Years")
sns.despine()
plt.show()
````

#### 4. In-depth Analysis
**Objective:** Conduct specific analyses to answer research questions

    Tasks:
    - just three type such as ```['Data Scientist', 'Data Engineer', 'Data Analyst']
    - Compare median salaries across different regions and company sizes.
    - Analyze the impact of experience level on salary in different job roles(just three type such as ```['Data Scientist', 'Data Engineer', 'Data Analyst']).

````python
for jt in dist_df['job_title'].unique():
    ch_d = dist_df[dist_df['job_title']==jt].groupby(['company_size','company_location'])[['salary_in_usd']].median().sort_values(by='salary_in_usd',ascending=False)
    ch_d.reset_index(inplace=True)
    fig, ax = plt.subplots(figsize = (12,6))
    alpha = 1
    for s in ch_d['company_size'].unique():
        ax.bar(x=ch_d[ch_d['company_size']==s]['company_location'],height=ch_d[ch_d['company_size']==s]['salary_in_usd'],alpha=alpha,label=s)
        alpha-=0.3

    ax.set_title("Salary of " + jt)
    ax.set_ylabel("Salary")
    ax.set_xlabel("Region")
    ax.legend()
    plt.show()
````

````python
for jt in dist_df['job_title'].unique():
    fig, ax = plt.subplots(figsize = (12,6))
    #sns.histplot(dist_df[dist_df['job_title']==jt], x='salary_in_usd', hue='', bins=50)
    sns.boxplot(dist_df[dist_df['job_title']==jt],x='experience_level', y='salary_in_usd')
    ax.set_xlabel("Salary of " + jt)
    ax.set_title("Salary Distribution by Experience Level")
    plt.show()
````
